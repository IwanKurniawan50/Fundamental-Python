# Scope dan LEGB

```python
kalimat = "ini merupakan Global"

def cetak():
    kalimat = "ini merupakan Local"
    
    def cetak1():
        kalimat = "ini adalah local bgt"
        print(f"saatnya dicetak {kalimat}")
        
    cetak1()
```

```python
cetak()
```

```python
angka = 2001

def nulis(angka):
    print(f"sebuah angka bernilai {angka}")
    
    angka = 2020
    print(f"namun sekarang berbubah menjadi {angka}")
```

```python
nulis(angka)
```

```python
botol = 666

def lotob():
    global botol
    
    botol = "xxx"
    print(f"pusing prennn ")
```
